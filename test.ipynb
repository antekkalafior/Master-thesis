```python
import json
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# List of filenames
filenames = ['training_history_1k.json', 'training_history_2k.json', 'training_history_3k.json', 'training_history_4k.json', 'training_history_5k.json']

# Initialize dictionaries to hold val_loss and val_accuracy for each model
val_losses = {}
val_accuracies = {}

# Load each file and extract 'val_loss' and 'val_accuracy'
for i, filename in enumerate(filenames):
    with open(filename, 'r') as f:
        data = json.load(f)
        val_losses[f'Model_{i+1}'] = data['val_loss']
        val_accuracies[f'Model_{i+1}'] = data['val_accuracy']

# Choose colormaps for blues (first 2 models) and greens (next 2 models)
blue_cmap = plt.get_cmap('Blues')
green_cmap = plt.get_cmap('Greens')

# Plot Validation Loss: first 2 models in blues, next 2 in greens, and last one in black
plt.figure(figsize=(10, 6))
for i, (model, val_loss) in enumerate(val_losses.items()):
    if i < 2:
        plt.plot(val_loss, label=model, color=blue_cmap((i + 1) / 3))  # Scalable blue color for first 2 models
    elif i < 4:
        plt.plot(val_loss, label=model, color=green_cmap((i - 1) / 3))  # Scalable green color for 3rd and 4th models
    else:
        plt.plot(val_loss, label=model, color='black')  # Black for the 5th model

plt.title('Validation Loss for All Models')
plt.xlabel('Epochs')
plt.ylabel('Validation Loss')

# Set x-ticks for every 1 epoch
num_epochs = len(val_loss)
plt.xticks(range(0, num_epochs, 1))

plt.legend()
plt.grid(True, which='both', axis='both')  # Show grid for both major and minor ticks
plt.savefig('Validation_loss_for_all_augmented.png', bbox_inches='tight')
plt.show()
plt.close()

# Plot Validation Accuracy: first 2 models in blues, next 2 in greens, and last one in black
plt.figure(figsize=(10, 6))
for i, (model, val_acc) in enumerate(val_accuracies.items()):
    if i < 2:
        plt.plot(val_acc, label=model, color=blue_cmap((i + 1) / 3))  # Scalable blue color for first 2 models
    elif i < 4:
        plt.plot(val_acc, label=model, color=green_cmap((i - 1) / 3))  # Scalable green color for 3rd and 4th models
    else:
        plt.plot(val_acc, label=model, color='black')  # Black for the 5th model

plt.title('Validation Accuracy for All Models')
plt.xlabel('Epochs')
plt.ylabel('Validation Accuracy')

# Set x-ticks for every 1 epoch
num_epochs = len(val_acc)
plt.xticks(range(0, num_epochs, 1))

plt.legend()
plt.grid(True, which='both', axis='both')  # Show grid for both major and minor ticks
plt.savefig('Validation_accuracy_for_all_augmented.png', bbox_inches='tight')
plt.show()
plt.close()
```

```python
import json
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# List of filenames
filenames = ['training_history_1k.json', 'training_history_2k.json', 'training_history_3k.json', 'training_history_4k.json', 'training_history_5k.json']

# Initialize dictionaries to hold val_loss and val_accuracy for each model
val_losses = {}
val_accuracies = {}

# Load each file and extract 'val_loss' and 'val_accuracy'
for i, filename in enumerate(filenames):
    with open(filename, 'r') as f:
        data = json.load(f)
        val_losses[f'Model_{i+1}k'] = np.array(data['val_loss'])
        val_accuracies[f'Model_{i+1}k'] = np.array(data['val_accuracy'])

# Choose colormaps for blues (first 2 models) and greens (next 2 models)
blue_cmap = plt.get_cmap('Blues')
green_cmap = plt.get_cmap('Greens')

# Plot Validation Loss: first 2 models in blues, next 2 in greens, and last one in black
plt.figure(figsize=(12, 6))
lines = []
labels = []
for i, (model, val_loss) in enumerate(val_losses.items()):
    if i < 2:
        color = blue_cmap((i + 1) / 3)  # Scalable blue color for first 2 models
    elif i < 4:
        color = green_cmap((i - 1) / 3)  # Scalable green color for 3rd and 4th models
    else:
        color = 'black'  # Black for the 5th model

    line, = plt.plot(val_loss, color=color)
    mean_loss = np.mean(val_loss)
    label = f'{model}: Mean Loss = {mean_loss:.4f}'
    lines.append(line)
    labels.append(label)

plt.title('Validation Loss for All Models')
plt.xlabel('Epochs')
plt.ylabel('Validation Loss')

# Set x-ticks for every 1 epoch
num_epochs = len(val_loss)
plt.xticks(range(0, num_epochs, 1))

# Add the legend in the top left corner
plt.legend(lines, labels, loc='upper left')
plt.grid(True, which='both', axis='both')  # Show grid for both major and minor ticks
plt.savefig('Validation_loss_for_all_augmented.png', bbox_inches='tight')
plt.show()
plt.close()

# Plot Validation Accuracy: first 2 models in blues, next 2 in greens, and last one in black
plt.figure(figsize=(12, 6))
lines = []
labels = []
for i, (model, val_acc) in enumerate(val_accuracies.items()):
    if i < 2:
        color = blue_cmap((i + 1) / 3)  # Scalable blue color for first 2 models
    elif i < 4:
        color = green_cmap((i - 1) / 3)  # Scalable green color for 3rd and 4th models
    else:
        color = 'black'  # Black for the 5th model

    line, = plt.plot(val_acc, color=color)
    mean_acc = np.mean(val_acc)
    label = f'{model}: Mean Accuracy = {mean_acc:.4f}'
    lines.append(line)
    labels.append(label)

plt.title('Validation Accuracy for All Models')
plt.xlabel('Epochs')
plt.ylabel('Validation Accuracy')

# Set x-ticks for every 1 epoch
num_epochs = len(val_acc)
plt.xticks(range(0, num_epochs, 1))

plt.legend(lines, labels, loc='lower right')
plt.grid(True, which='both', axis='both')  # Show grid for both major and minor ticks
plt.savefig('Validation_accuracy_for_all_augmented.png', bbox_inches='tight')
plt.show()
plt.close()
```
